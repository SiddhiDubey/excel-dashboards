# Data Cleaning

```python
netflix_copy_df['director'].fillna('No Director', inplace=True)
netflix_copy_df['cast'].fillna('No Cast', inplace=True)
netflix_copy_df['country'].fillna('Country Unavailable', inplace=True)
netflix_copy_df['rating'].fillna('Rating Unavailable', inplace=True)
netflix_copy_df['duration'].fillna('Duration Unavailable', inplace=True)
netflix_copy_df.dropna(subset=['date_added'],inplace=True)
```

## 8.Number of Movies and TV Shows added to Netflix (by year

```python
netflix_copy_df['year_added'] = netflix_copy_df['date_added'].str.slice(start=-4)
netflix_copy_movies_and_tvshows = netflix_copy_df.groupby(['year_added', 'type'], as_index=False).show_id.count()
netflix_copy_movies_and_tvshows.columns = ['year_added', 'type', 'count']
netflix_copy_movies_and_tvshows = netflix_copy_movies_and_tvshows.pivot(index='year_added',
                                                              columns='type',
                                                              values='count').reset_index()
netflix_copy_movies_and_tvshows.fillna(0, inplace = True)

# Create a DataFrame called netflix_copy_data that contains the year, number of movies, and number of TV shows
netflix_copy_data = pd.DataFrame(dict(Year   = netflix_copy_movies_and_tvshows['year_added'],
                                      Movie  = netflix_copy_movies_and_tvshows['Movie'],
                                      TVshow = netflix_copy_movies_and_tvshows['TV Show']))

# Display the dataset for the new netflix_copy_data Dataframe
netflix_copy_data
```

```python
ind = np.arange(len(netflix_copy_data))
width = 0.36

fig1, ax1 = plt.subplots()
ax1.barh(ind, netflix_copy_data.Movie, width, color='#3399FF', label="Movies")
ax1.barh(ind + width, netflix_copy_data.TVshow, width, color='#9966CC', label="TV Shows")

ax1.set(yticks=ind + width, yticklabels=netflix_copy_data.Year, ylim=[2*width - 1, len(netflix_copy_data)])
ax1.set_title("Number of Movies and TV Shows added to Netflix (by year)", fontsize=18)
ax1.set_ylabel("Year added to Netflix", fontsize=14)
ax1.set_xlabel("Number of Movies and TV Shows", fontsize=14)
ax1.legend()

fig1.set_size_inches(16, 16)
plt.savefig('Number_of_Movies_TV_DF1.png')
plt.show()
```

## 9.Top 5 Durations Based on The Number of Titles

```python
plt.figure(figsize=(10,8))
netflix_duration = netflix_copy_df['duration'].value_counts()
netflix_duration = pd.DataFrame(netflix_duration).reset_index()
netflix_duration.columns = ['duration','title']
sns.barplot(x = 'duration',y = 'title', data=netflix_duration.head(5), palette="cividis_r")
plt.title('Top 5 Durations Based on The Number of Titles', fontsize=21);
```
